```python
import pandas as pd
import os
import time
from google import genai
from google.genai import types

# Carregar seu arquivo CSV
client = genai.Client(api_key=os.getenv("dio-api-dados"))
```

```python
# O pandas vai ler os IDs 1, 2, 3, 4 
try:
    df = pd.read_csv('desafio.csv') 
    user_ids = df['User ID'].tolist()
except Exception as e:
    print(f"Erro ao ler o arquivo: {e}")
    user_ids = []
resultados = []
```

```python
# Loop
for uid in user_ids:
    try:
        
        response = client.models.generate_content(
            model="gemini-2.5-flash", 
            contents=f"Crie uma frase curta de investimento para o cliente de ID {uid}. Máximo 50 caracteres."
        )
        
        texto = response.text.strip()
        resultados.append({"id": uid, "mensagem": texto})
        print(f"ID {uid} processado com sucesso!")
        
        # IMPORTANTE: A conta gratuita tem limite por minuto. 
        time.sleep(4) 
        
    except Exception as e:
        if "429" in str(e):
            print(f"Limite atingido no ID {uid}. Aguardando 15 segundos...")
            time.sleep(15)
            # Tenta novamente uma vez após a espera
            response = client.models.generate_content(model="gemini-2.5-flash", contents=f"Investimento para ID {uid}")
            resultados.append({"id": uid, "mensagem": response.text.strip()})
        else:
            print(f"Erro no ID {uid}: {e}")
            resultados.append({"id": uid, "mensagem": "Erro técnico"})

# Criar o DataFrame final e exibir
df_final = pd.DataFrame(resultados)
display(df_final)

# 5. Salvar o resultado
df_final.to_csv('resultado_marketing.csv', index=False, encoding='utf-8')
print("\nArquivo 'resultado_marketing.csv' salvo com sucesso!")
```
